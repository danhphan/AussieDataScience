```python
import pandas as pd
from parsel import Selector
from selenium import webdriver
from selenium.webdriver import *
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time
```

```python
PATH = '/home/danph/Repos/run/drivers/chromedriver'

def get_driver(url):
    chrome_options = Options()
    chrome_options.add_argument("--headless")
    driver = webdriver.Chrome(
        executable_path=PATH, 
        options=chrome_options)
    driver.get(url)
    return driver
```

```python
URL = "https://www.linkedin.com/jobs/search/?keywords=data%20scientist&location=Australia&sortBy=DD"
driver = get_driver(URL)
```

```python
def get_job_info(selector):
    title = selector.css('h3::text').get()
    location = selector.xpath("//span[contains(@class,'card__location')]/text()").get()
    link = selector.xpath("//a/@href").get()    
    print(title, location, link)    
```

```python
job_list = driver.find_elements_by_xpath("//ul[contains(@class,'jobs-search__results-list')]/li")
for job in job_list:
    time.sleep(0.2)
    print("==========")
    selector = Selector(text=job.get_attribute("outerHTML"))
    get_job_info(selector)
```
